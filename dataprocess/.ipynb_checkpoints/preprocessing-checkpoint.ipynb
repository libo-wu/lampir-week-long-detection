```python
# create namelist from filelist, just extract/seperate the file name of LAMPIR files and PIR files
def create_namelist(filelist):
    lampirlist=[]
    pirlist=[]
    for filename in filelist:
        if(filename[-3:]=='csv'):
            if(filename[-7:]=='pir.csv'):
                pirlist.append(filename)
            else:
                lampirlist.append(filename)
        else:
            continue
    
    return lampirlist, pirlist
```

```python
# plot 1 minute samples ~4800 with time stamps on xaxis.
# setxtick='1': Set ticks labels for x-axis with time stamps, slow
# setxtick='0': not set x ticks
def showplot(arr,setxtick):
    num_samples=arr.shape[0]
    y=arr[:,1]
    x=arr[:,0]
    x_tick=[]
    for i in range(0,arr.shape[0]):
        if(i%800==0):
            time_stamp=str(arr[i][3])+'-'+str(arr[i][4])+'-'+str(arr[i][5])+'-'+str(arr[i][6])+'-'+str(arr[i][7])
            x_tick.append(time_stamp)
        else:
            x_tick.append(' ')
    if (setxtick=='1'):
        fig, ax = plt.subplots(1,1) 
        ax.plot(x,y)
        ax.set_xticks(x)
        # Set ticks labels for x-axis
        ax.set_xticklabels(x_tick, rotation='vertical', fontsize=12)
        ax.set_ylim(0,1023)
    else:
        plt.plot(x,y)
        plt.ylim(0,1023)
        plt.show()
```

Input: k-window size, i-index, T-signal series
\begin{equation*}
 Score(k,i,T)=\frac{max(x_i-x_{i-1},x_i-x_{i-2},...,x_i-x_{i-k})+max(x_i-x_{i+1},x_i-x_{i+2},...,x_i-x_{i+k})}{2}
\end{equation*}

```python
# function Compute_Score(k,i,T)
# input: k-window size, i-index, T-signal series
def Compute_Score(k,i,T):
    N=len(T)
    left=[]
    right=[]
    if (i<k):
        right=T[i:(i+k)] 
        left=[0]*k
    elif (i<N-k):
        left=T[(i-k):i]
        right=T[i:(i+k)]
    else:
        left=T[(i-k):i]
        right=[0]*k
    value=(np.max(left)+np.max(right))/2
    return value
```

```python
# find peaks in time series
# <Simple Algorithms for Peak Detection in Time-Series> by Girish Keshav Palshikar
def findpeak(T,k,h):
    peaklist=[]
    N=len(T)
    score=[]
    for i in range(0, N):
        score.append(Compute_Score(k,i,T))
    pos_score=[s for s in score if s>0]
    mean=np.mean(score)    #compute positive values in score
    std=np.std(score)
    for i in range(0,N):
        if (score[i]>0 and (score[i]-mean)>h*std):
            peaklist.append(i)
    peaklist.sort()
    # retain only one peak out of any set of peaks within distance k of each other
    
    return peaklist

%matplotlib inline
plt.rcParams['figure.figsize'] = (20.0, 10.0) # set default size of plots

arr=datalist[22]
y=arr[:,1]
x=arr[:,0]
peaklist=findpeak(y,10,1.7)
peakvalue=y[peaklist]
plt.subplot=(2,1,1)
plt.plot(x,y)
plt.ylim(0,1023)
plt.plot(2,1,2)
plt.plot(peaklist,peakvalue,'ro')
plt.ylim(0,1023)

plt.show()
```
