```python
# import modules
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import re
import csv

```

```python
# create namelist from filelist, convert '2018-03-08 17-57-38.csv' to '2018_03_08_17_57_38'
def create_namelist(filelist):
    namelist=[]
    for filename in filelist:
        name=re.split('\D',filename)
        if(name[0]=='2018' or name[0]==2019):   #this is trival decide strategy
            newname='lampir_'+name[0]+'_'+name[1]+'_'+name[2]+'_'+name[3]+'_'+name[4]+'_'+name[5]
        else:
            continue
        namelist.append(newname)
    
    return namelist
```

```python
# plot 1 minute samples ~4800 with time stamps on xaxis.
# setxtick='1': Set ticks labels for x-axis with time stamps, slow
# setxtick='0': not set x ticks
def showplot_1min(arr,setxtick):
    num_samples=arr.shape[0]
    y=arr[:,1]
    x=arr[:,0]
    x_tick=[]
    for i in range(0,arr.shape[0]):
        if(i%800==0):
            time_stamp=str(arr[i][3])+'-'+str(arr[i][4])+'-'+str(arr[i][5])+'-'+str(arr[i][6])+'-'+str(arr[i][7])
            x_tick.append(time_stamp)
        else:
            x_tick.append(' ')
    if (setxtick=='1'):
        fig, ax = plt.subplots(1,1) 
        ax.plot(x,y)
        ax.set_xticks(x)
        # Set ticks labels for x-axis
        ax.set_xticklabels(x_tick, rotation='vertical', fontsize=12)
        ax.set_ylim(0,1023)
    else:
        plt.plot(x,y)
        plt.ylim(0,1023)
        plt.show()
```

```python
#plot func to visualize voltage signal with time 
#input: `datalist`, type=list, size=num_arrays*num_samples_1min*9
#output: a subplot figure shows all arrays in this list, each array per subplot 
#title is the start time of this array(1min)
def showplot_list(datalist):
    num_arrays=len(datalist)
    num_samples_list=[]
    all_samples=[]
    plot_cols=2
    plot_rows=int(num_arrays/2)+num_arrays%2
    for i in range(0,num_arrays):
        plt.subplot(plot_rows,plot_cols,i+1)
        arr=datalist[i]
        y=arr[:,1]
        x=arr[:,0]
        plt.plot(x,y)
        plt.ylim(0,1023)
        title=str(arr[0][2])+'-'+str(arr[0][3])+'-'+str(arr[0][4])+'-'+str(arr[0][5])+'-'+str(arr[0][6])+'-'+str(arr[0][7])
        plt.title(title)
    
    plt.show()
```
